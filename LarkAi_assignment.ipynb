```python
import cv2
import tensorflow as tf
```

```python
image_path = "/Users/suhaylmahek/Downloads/assignment folder/4.jpg"
image = cv2.imread(image_path)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
image_tensor = tf.convert_to_tensor(image_gray, dtype=tf.float32)
```

```python
image_tensor /= 255.0
```

```python
print(image_tensor.shape)
```

```python
ans=[]
```

```python
for i in range(len(image_tensor[0])):
    array=[]
    for j in range(len(image_tensor)):
        if image_tensor[j][i]==0.0:
            array.append(float(0))
        else:
            array.append(float(1))
    arg=np.argmin(array)
    arg=120-arg
    if arg>=117:
        ans.append(ans[-1])
    else:
        ans.append(arg)

            
            
```

```python
len(ans)
```
